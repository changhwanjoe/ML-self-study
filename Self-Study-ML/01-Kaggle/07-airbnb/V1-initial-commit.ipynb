```python
def convert_date(timestring):
    timestring = str(timestring)
#     timestring[:4]), (timestring[4:6]), (timestring[6:8]), (timestring[8:10]), (timestring[10:12]), (timestring[12:14])
    return "{}-{}-{} {}:{}:{}".format(timestring[:4], timestring[4:6], timestring[6:8], timestring[8:10], timestring[10:12], timestring[12:14])
```

```python
all_data['timestamp_first_active'] = pd.to_datetime(all_data['timestamp_first_active'].apply(convert_date))
```
