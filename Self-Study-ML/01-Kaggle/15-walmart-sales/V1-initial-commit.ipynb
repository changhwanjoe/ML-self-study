```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

plt.rcParams["figure.figsize"] = (18,8)
```

```python
train = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/train.csv', parse_dates=['Date'])
test = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/test.csv', parse_dates=['Date'])
stores = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/stores.csv')
sampleSubmission = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/sampleSubmission.csv')
```

```python
train['year'] = train['Date'].dt.year
train['month'] = train['Date'].dt.month
train['day'] = train['Date'].dt.day
train['week'] = train['Date'].dt.week
```

```python
train['num_week'] = np.ceil(train['day'] / 7)
```

```python
train['num_week'].value_counts()
```

```python
test['year'] = test['Date'].dt.year
test['month'] = test['Date'].dt.month
test['day'] = test['Date'].dt.day
test['week'] = test['Date'].dt.week
```

```python
test['num_week'] = np.ceil(test['day'] / 7)
```

```python
test['num_week'].value_counts()
```

```python
train.head()
```

```python
stores['Code'] = stores['Type'].astype('category').cat.codes
```

```python
train = pd.merge(train, stores, how='left')
```

```python
test= pd.merge(test, stores, how='left')
```

```python
train.head()
```

```python
test.head()
```

```python
train['Weekly_Sales'].describe()
```

```python
misc = 1100
```

```python
train_1 = train[train['Weekly_Sales'] >-misc]
```

```python
train_1.sort_values(by='Weekly_Sales', ascending=False).tail(30)
```

```python
train_1['Weekly_Sales_log'] = np.log1p(train_1['Weekly_Sales'] + misc)
```

```python
train_1.head()
```

## Make Prediction

```python
feature = ['Store', 'Dept', 'IsHoliday', 'month', 'year', 'day', 'Code', 'Size', 'week', 'num_week']
label = ['Weekly_Sales_log']
```

```python
# from sklearn.ensemble import RandomForestRegressor
```

```python
# rfr = RandomForestRegressor(n_estimators=100, random_state=30, n_jobs=4)
```

```python
from lightgbm import LGBMRegressor
```

```python
# rfr = LGBMRegressor(colsample_bytree=0.8, subsample=0.8, num_leaves=4000, n_estimators=130, reg_lambda=4, reg_alpha=0.5)
```

```python
rfr = LGBMRegressor(colsample_bytree=0.8, 
                    subsample=0.8, 
                    num_leaves=4000, 
                    n_estimators=520, 
                    min_child_weight=0.05, 
                    class_weight='balanced',
                    importance_type='gain',
                    objective='regression',
                    learning_rate=0.025,
                    )
```

```python
rfr
```

```python
x_train = train_1[feature]
y_train = train_1[label]
x_test = test[feature]
```

```python
rfr.fit(x_train, y_train)
```

```python
prediction = rfr.predict(x_test)
```

```python
prediction = np.expm1(prediction)-misc
```

```python
rfr.feature_importances_
```
